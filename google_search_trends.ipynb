## Google Search Trends

### 1.Overview

Every day, Google receives around 3.5 billion searches, equivalent to roughly 40,000 searches per second. This makes Google search an excellent opportunity for analyzing search query data.

### 2.Target

Use the Google Trends API to make it easy to analyze daily Google searches.

### 3.DataSet information:

We will utilize the Google Trends API (pytrends) to construct a dataset containing the top 10 countries that search for "Artificial Intelligence" on Google

### 4.Import libraries

```python
import pandas as pd
# imports the TrendReq class from the pytrends.request module
# allows you to make requests and retrieve data from Google Trends.
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
# creates an instance of the TrendReq class and assigns it to the variable "trends"
# make requests and retrieve data from the Google Trends API.
trends = TrendReq()
```

### 5.Import DataSet

```python
#Create DataFrame of the top 10 countries which search for “Artificial Intelligence” on Google:
trends.build_payload(kw_list=["Artificial Intelligence"])
df_ai = trends.interest_by_region()
df_ai = df_ai.sort_values(by="Artificial Intelligence", ascending=False)
df_ai=df_ai.head(10)
print('Top Countries Searching for "Artificial Intelligence":\n',df_ai)
```

**OBS**:

* So, according to the above results, the search queries based on "Artificial Intelligence" are mostly done in Ethiopia.

### 6.Data Visualization:

```python
#Create a bar chart:
df_ai.reset_index().plot(x="geoName", 
                       y="Artificial Intelligence", 
                       figsize=(10,8), kind="bar")
                       
plt.style.use('dark_background')
plt.show()
```

As we all know, Artificial Intelligence has caught the attention of numerous companies and students in recent years. 

Let's explore the search trend to see how queries on Google have evolved in recent years.

```python
#Create a new Dataframe for searches related to artificial intelligence in recent years
#h='en_US':Language we use to connect to Google Trends
#tz=360 : Time Zone CST (Central Standard Time)
df_ai_time = TrendReq(hl='en-US', tz=360)
df_ai_time.build_payload(kw_list=['Artificial Intelligence'])
df_ai_time = df_ai_time.interest_over_time()
```

```python
# create a linechart "Total search" vs "Year":
fig, ax = plt.subplots(figsize=(10, 8))
df_ai_time['Artificial Intelligence'].plot()
plt.style.use('dark_background')
plt.grid(visible=True,color='r', linestyle='dashed', linewidth=0.6)
plt.title('Total Google Searches for Artificial Intelligence', 
          fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Search Count')
plt.show()
```

**OBS:**

Based on the previous chart:
* we can observe a significant increase in searches for "Artificial Intelligence" on Google since the end of 2022.
* In the early months of 2023, the trend experiences a remarkable surge.

### 7.Final Observations:
